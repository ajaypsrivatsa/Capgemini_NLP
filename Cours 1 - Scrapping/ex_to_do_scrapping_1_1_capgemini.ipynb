## 1. Make necessary imports

```python
%pip install scrapy
import requests
from scrapy.selector import Selector
```

## 2. Get some information from Capgemini Invent's website

In order to discover what selectors are, you will extract some basic information from [this page](https://www.capgemini.com/service/invent/)

```python
# Define  the url
url = 'https://www.capgemini.com/service/invent/'
```

```python
# Create selector object
html = requests.get(url).content
sel = Selector(text=html)
```

## 3. Let's start with a basic example : extracting Capgemini Invent's moto with selectors

_As a reminder, you can use `F12` to inspect a webpage from your favorite navigator_

```python
# Getting the Capgemini Invent moto thanks to css selector
css_locator = 'h1 ::text'
rep = sel.css(css_locator).extract_first()
print(rep)
```

```python
# Doing the same with xpath selector
xpath_locator = '//h1/text()'
rep = sel.xpath(xpath_locator).extract_first()
print(rep)
```

## 4. Now it is your turn, here are some exercices for you

__<font color='Blue'>
Exercice 1 : Using selectors, get the name of all the Capgemini Invent experts
</font>__

```python
xpath_locator = '//h3[@class = "card_person__title"]/a/text()'
rep = sel.xpath(xpath_locator).extract()
print(rep)
```

__<font color='Blue'>
Exercice 2 : Using selectors, get the description of these experts that you found
</font>__

```python
xpath_locator = '//div[@class = "card_person__text"]/p/text()'
rep = sel.xpath(xpath_locator).extract()
print(rep)
```

__<font color='Blue'>
Exercice 3 : Using selectors, get the name of all the Capgemini Invent capabilities
</font>__

```python
trans_table = {ord(c): None for c in u'\r\n\t'}
xpath_locator = '//div[@class="card_tab__hover"]/h3[@class = "card_tab__title"]/text()'
rep = [', '.join(s.translate(trans_table) for s in sel.xpath(xpath_locator).extract())]
print(rep)
```
