## Preparing selectors in spider perspective

### 1. Parsing restaurant list

```python
html = requests.get(example_url_city).content
sel = Selector(text=html)
resto_list = []
i = 1
while sel.xpath('//a[@class ="nav next rndBtn ui_button primary taLnk"]/@href').extract_first() is not None :
  next_url = 'https://www.tripadvisor.co.uk' + sel.xpath('//a[@class ="nav next rndBtn ui_button primary taLnk"]/@href').extract_first()
  xpath_locator = '//div[@class = "wQjYiB7z"]/span/a/text()'
  rep = sel.xpath(xpath_locator).extract()
  for s in rep :
    if s == '. ' or s.isdigit() :
     continue
    else :
      resto_list.append(s)
  html = requests.get(next_url).content
  sel = Selector(text=html)
  i += 1
xpath_locator = '//div[@class = "wQjYiB7z"]/span/a/text()'
rep = sel.xpath(xpath_locator).extract()
for s in rep :
  if s == '. ' or s.isdigit() :
    continue
  else :
    resto_list.append(s)
print(resto_list)
```
